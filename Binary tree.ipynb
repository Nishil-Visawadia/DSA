```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
        
        
class BT:
    def  __init__(self,data):
        self.root = Node(data)      
        
    def preorder(self,root):
        if root == None:
            return
        
        print(root.data,end=' ')
        self.preorder(root.left)
        self.preorder(root.right)
        
    def inorder(self,root):
        if root == None:
            return
        
        self.inorder(root.left)
        print(root.data,end=' ')
        self.inorder(root.right)
        
    def postorder(self,root):
        if root == None:
            return
        
        self.postorder(root.left)
        self.postorder(root.right)
        print(root.data,end=' ')   
        
    def countNodes(self,root):
        if root == None:
            return 0
        
        return self.countNodes(root.left) + self.countNodes(root.right) + 1
    
    def SumNodes(self,root):
        if root == None:
            return 0
        
        return self.SumNodes(root.left) + self.SumNodes(root.right) + root.data
    
    
    def HeightBT(self,root):
        if root == None:
            return 0
        
        return max(self.HeightBT(root.left),self.HeightBT(root.right)) + 1
    
    
    def SumReplace(self,root):
        if root == None:
            return None
        
        
        self.SumReplace(root.left)
        self.SumReplace(root.right)

        if root.left != None:
            root.data += root.left.data
            
        if root.right != None:
            root.data += root.right.data
            
            
        
tree = BT(1)
tree.root.left = Node(2)
tree.root.right = Node(3)
tree.root.left.left = Node(4)
tree.root.left.right = Node(5)
tree.root.right.left = Node(6)
tree.root.right.right = Node(7)
tree.root.right.right.right = Node(8)

tree.preorder(tree.root)
print()
tree.inorder(tree.root)
print()
tree.postorder(tree.root)
print()
print(tree.countNodes(tree.root))
print(tree.SumNodes(tree.root))
print(tree.HeightBT(tree.root))
tree.SumReplace(tree.root)
tree.preorder(tree.root)
```
